There are various challenges associated with Database.
1.The database might be huge, with tens of thousands of tables in one database.
2.And some of those tables may have millions or even billions of rows in a single table, imagine that.
3.Queries running against a table with billions of rows, those queries are going to take a long time, even moving and copying the data can take a long time.
4.I could have thousands or even millions of users hitting the same database table at the same time. How does the software keep track of that? 
5.How about privacy and security? How can the database software keep one user's work from clobbering another user's work? 

TRANSACTION LOGtransaction logs are the method that database software uses to make sure that the database remains available. And uncorrupted and safe for many users to use as we run our various applications, transaction logs.



BUFFERCACHE : Anything that's read from the database or written back to the database resides in buffer cache in memory on the database server. So I may read a row from a table, it goes into the buffer cache and then I can update it and write it back out, it's written from the buffer cache back out to the disk.
