However, in case we need to analyze data, rather than collect data, we need to create a database for the subject of interest to support decision making. In this module, we are going to learn a different perspective of organizing the data, Data Warehouse, or Online Analytical Processing.

WHAT DO WE NEED TO ORGANIZE A DATA?
1. SCHEMA- How should my data be logically organized.
2. NORMALIZATION - Should my data have normal dependency( thus redudancy)
3. VIEWS - What joins will be done most often?
4. ACCESS CONTROL - Should all users of data have same level of access?

So when we say a different Perspective of Organizing Data we mean a Data Warehouse.

## DATAWAREHOUSE: 
The data warehouse is actually a physically separating the store of data and it is transformed from a operational environments. There are two different perspectives. One is for the operational environments, we have to do a lot of data updating and keep it different, and for the data warehouse, we want to make a static view over it just for a time period so we can deal with those data in a data warehouse. That's what reflects another perspective, that is, data warehouse may not have up-to-date data. We only have a certain period of time data loaded into the data warehouse, and then there's all and then we are going to wait for another period, a month, a quarter, a year to come and then we load the new data. There is no operation update of data at all in the data warehouse, at least not daily, not weekly, maybe just monthly or yearly, so that we don't have to take care of the transaction processing, because we do not do transaction and we do not emphasize our recovery because those data are converted from operational environments.
