#### Database management system ####
This is a way for us to organize our data.If we do not make a comparison with multiple pieces of information, the data actually are not that valuable as we thought. We need a very systematic efficient way to manage our data, so that whenever we want to make some decisions we can turn to our data and let our data speak.
We try to find out all piece of information when they come together, what patterns we can find out, and that will be our knowledge. Then starting from knowledge, we're going to try to find correlation and even causality. Then that will give us some understanding of the principles. These principles will finally give us the intelligence or wisdom that can guide us to make decision. That will be the data, and then that will be the actual valuable data, rather than just a piece of information.
