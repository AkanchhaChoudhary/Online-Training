### Why Distributed Computing? ###

To understand this ,first we have to understand what a big data is.

Businesses are seeing the value of data, and the size of that data is growing. Better use of data means that we can make better, more accurate, and more automatic decisions. Why rely on hunches or the way we've always done things when data is available? To put some rough numbers around it, collectively, we're generating 2.5 million terabytes of data per day. Total data storage will be 175 zettabytes, that's 175 billion terabytes by 2025. Big data generally refers to a variety of different qualities known as the five V's. The five V's are volume, velocity, variety, veracity, and value. First and foremost, the size of our data is growing. This is the volume component of big data. The second way that data is growing is in velocity. As we have more and more mechanisms that can produce data, it is arriving in data pipelines at a faster and faster rate. We are also seeing increases in the variety of data that we have to deal with, ranging from numeric and textual data to images to video streams, etc. There's also the question of veracity or how much can we trust the data that we have? Some data arriving in our systems may have missing values or may be inaccurate, or maybe it's crowdsource, so it can't always be trusted. Finally, data is highly valuable, because it allows us to make better, faster decisions. At a high level, big data is generally referred to as more data that can fit on any single machine. This can range from gigabytes up to terabytes and petabytes and beyond. Think about maxing out your hard drive. If you fill up the storage on your laptop and you still have data left to put somewhere, how much you process it. This is where distributed computing comes in. 




### Why Spark ###
To deal with such large data we need to divide that data and the processing across a number of different machines in order to be able to interact with it. That's where Spark comes in. Spark solution to the problem of big data is to provide the first unified analytics engine. 
Spark simplifies working with data by supporting different languages, including SQL, a common language among data analysts. It supports Python and R for data scientists, and Scala and Java, which is common among software engineers. Using any of these languages, data analysts can run queries and reports, data scientists can train and deploy machine learning models, and data engineers can use Scala to work with live streams of incoming data and clean new data in nightly batch drops. Spark also rose to popularity by reading and processing data from many disparate sources, including databases like Postgres or MySQL, Blob Stores like Amazon's S3, data warehouses and files like CSVs.Spark is the open-source industry standard tool for manipulating big data.The two main use cases for Spark, are when you need to scale out or speedup. You need to scale out if you have too much data to process on a single machine. Even if you're data can fit on a single machine, you can still benefit from speeding up your query by adding more compute resources.

### Spark DataFrames ###
When Spark was first introduced, it only had the RDD API. A few years later, the DataFrame API was introduced. This provided additional optimizations by adding metadata or schema on top of these underlying RDDs. It also introduced new programming languages such as R and SQL. DataFrame actually inherits a lot of those properties of RDDs plus they add additional metadata. This metadata includes the number of columns and data types, and these are highly optimized.Spark intentionally decouples storage and compute, but it's able to operate on this data across multiple nodes or executors.Because the underlying optimizer and Spark catalyst understand what type of actions are transformations you're trying to apply to your data set, it's able to highly optimized the query plan that gets generated, and so you should always use the DataFrame API wherever possible. If you're using Spark SQL, it always executes against a Spark DataFrame. Spark is not a database. It's a compute engine that can read from databases, but as I just mentioned, it decouples storage and compute. The data is ephemeral. What that means is that if your Spark cluster goes down, you don't actually lose any of the underlying data. You might lose whatever data you are currently processing, but the source data is still present. Just whatever intermediate data you are computing might not be present if the cluster goes down partway through processing. 
The DataFrame API is declarative, you specify what you want to be done, whereas with the RDD API, it was imperative, you specify how you wanted something to be done, and so there wasn't any optimization to be done.Spark dataframe uses internal optimizer called catalyst. With catalyst, you specify what you want done rather than how you want it done. 

### Databrics ###
Databricks is a unified analytics platform that enables data science and engineering teams to run all analytics in one place. It works on the three main Cloud providers, Amazon, Microsoft, and Google. You can work with data in any format whether it's structured, semi-structured, unstructured, or streaming. It's powered by many open-source tools. This allows you to run reports and power dashboards, as well as run, extract, transform, load jobs where new data is cleaned and inserted into databases.Databricks is a more optimized managed version of the open-source Apache Spark project, offering some key benefits over basic Spark. The Databricks platform consists of a few different components. Most important to us in this course is the hosted notebook environment. This means that we can interact with our data in real time by running selves of code hosted on a Databrick server.
