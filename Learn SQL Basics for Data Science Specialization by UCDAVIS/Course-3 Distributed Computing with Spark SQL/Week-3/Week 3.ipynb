SPARK AS A CONNECTOR
Below in the image we can see a common big data architecture



Data is landed in a raw format, oftentimes using Apache Kafka or some Apache Kafka-like technology.The raw data is then parsed out through an ETL process. This is the process where we take raw data, transform it, and then load it as a clean form in some database. Afterwards, our data is stored in various places and formats for ad hoc queries, for use in databases, for reporting applications in machine learning model training. Like I said, there are many different architectures out there and loads of different technologies to achieve the needs of data applications.A data application needs a scalable way to receive high volume velocity and variety data. This means that when data arrives in our system, it won't overload other aspects of our infrastructure. This piece of our infrastructure operates as a queue. The most common solution for this queue is the technology Apache Kafka, another open source project. Spark offers an optimized compute engine that connects to data where it lives. This means that data can be sitting in a number of different storage technologies, from various databases to massive objects stores to streaming engines. One core design principle is therefore, decoupled storage and compute.By spinning up a Spark cluster when it's needed, connecting to data where it lives and shutting it down afterwards, not only can you save resources, but you don't have to worry about updating your cluster to the latest version of Spark. Version updates can cause significant challenges for older generation software. Spark also leverages data stores for what they're good at, storing and serving data and leverages Spark for what it's good at, running optimized computation across that data. There are generally two types of bottlenecks on computing tasks. IO and CPU bound problems.

JSON, Schemas and Types
When we think about data, the most common association is tabular data. This means ordered rows and columns of data like in a sequel table or excel file. This works for many use cases. But what happens when your data changes or you want to add a new column? You'll end up with no values where you're missing data or what about complex types? Like having a column as a list of different values? Yes, you could do some of this using tabular data, but it's not quite the best tool for the job. This is why in many applications you find semi structured data instead. Semi structured data does not conform to a formal structure like columns and rows. Fields are not necessarily shared by all of the observations. This approach is also known as a self describing structure. This allows for better nested values and hierarchies.

WRITING DATA.
Data writes can be parallelized by changing the number of partitions of our data. This means that our writes will be faster, and when we read from that data, we can read in parallel as well. By the end of this lesson, you will have written to a target database in both serial and in parallel. Let's start by talking a little bit about partitions. Partitions control the reads and writes that we perform. We're also going to control these partitions using the repartition and coalesce hints. About partitions. Due to Spark's distributed nature, database rights work a little bit different from what you've seen in the past. Now, a partition is a portion of our total data set. Recall that when we interact with data frames in Spark, in reality, our data is distributed across the cluster. It's not just all sitting together in memory on one machine. A partition is one of those portions of our data set. When we read or write to a target database, we're going to have one connection for each of the different partitions that we have. This is the main way we control parallelism with our writes from Spark. Below is the link to databrics notebook for more information to understand the data writing.
