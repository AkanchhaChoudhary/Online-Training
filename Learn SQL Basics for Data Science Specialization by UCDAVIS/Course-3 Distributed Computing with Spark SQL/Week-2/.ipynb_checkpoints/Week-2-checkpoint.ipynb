Its good to know specific Spark terminology such as partitions and slots that will help you optimize your queries.A Spark cluster is made up of one driver and one or more executors. But if you want to keep cost down and prototype the Spark, you can use something called spark local mode, in which case the driver and the executor are on the same physical machine. In this case, when you're trying to do any Spark queries, the driver tells the executor what to do, then turns around and actually does the work itself. Spark operates on a cluster of machines. Each of those machines can have multiple cores and each of those cores can have multiple threads. We denote the smallest unit of parallelism, whether it is a core or thread as a slot. In this diagram, you can see that there are four executors, each with two slots, which makes the unit of parallelism eight. That's a lot to keep track of. Think about all of these terms, machines, cores, threads as some unit of parallelism. Just because you have slots available, doesn't mean that Spark will necessarily use them.

### SHUFFLE PARTITIONS: ###
This is one of my favorite topics because customers love to ask how can I make my spark sequel query run faster and oftentimes they think we're going to tune their, select their from their wear. But in reality it often comes down to one of two main things. One is their underlying data layout on disk. The second part are these default spark configurations such as shuffle partitions .Transformations are significantly more expensive to run because they force a shuffle of data across all of the executors.Narrow transformations include select drop where etc. To compute a narrow transformation all of the data needed resides within a single partition, for example. To select a subset of columns I don't need to look at records within other partitions to be able to accomplish that task this is in contrast with wide transformations. With wide transformations such as distinct group by your order by you can compute a subset of it locally. But you're not going to get the global values unless you aggregate them across all of the different partitions. So here's an example of a group by, I can group by all the reds locally, the greens, the blues. But we need some way of being able to group them across all of these different partitions and so this is a wide transformation.The Spark SQL shuffle is a mechanism for redistributing or re-partitioning data so that the data is grouped differently across partitions, based on your data size you may need to reduce or increase the number of partitions of RDD/DataFrame using spark.shuffle partitions configuration can have a big impact on the runtime performance of our query. However, the big challenge of shuffle partitions is there is no universal partition number because data changes at different times of query execution. we can set the shuffle partitions configuration multiple times even within a single notebook.

### Adaptive Query Execution (AQE) ###
AQE is a new feature introduced in spark three auto to automatically update your query plans based on runtime statistics.AQE provides is adaptive planning. It makes decisions based on accurate runtime statistics and automatically happens whenever there is a stage boundary which is an inherent breakpoint. It does dynamic query optimization that happens in the middle of query execution based on runtime statistics. It will continually check when the dependencies are finished, it will go ahead and reoptimize the unexecuted part of the query and if there are more stages to run, it continues this process until all of them have been completed. There are three key features that AQE provides. Dynamically coalescing shuffle partitions, dynamically switching joint strategies, and dynamically optimizing skew joints. We'll dive into each of these three individually.AQE tackles the problems of shuffle partitions problem by setting the initial partition number to be high but will automatically coalesce these partitions.
